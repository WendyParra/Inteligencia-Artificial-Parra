```python
import tensorflow as tf # Importación de libreria
import numpy as np # Importación de libreria

# Datos de entrada
anios = np.array([2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015], dtype=float) #valores de entrada

# Datos de salida
calidad_buena = np.array([77, 148, 164, 105, 156, 75, 84, 175, 127, 41], dtype=float)  #valores de salida esperados
calidad_regular = np.array([166, 201, 180, 217, 183, 284, 249, 173, 203, 235], dtype=float)  #valores de salida esperados
calidad_mala = np.array([0, 2, 1, 0, 12, 6, 260, 17, 35, 70], dtype=float)  #valores de salida esperados

# Definir el modelo
modelo = tf.keras.Sequential([ #definir secuencia de  capas
    tf.keras.layers.InputLayer(input_shape=[1]),  # capa de una sola entrada
    tf.keras.layers.Dense(units=64, activation='relu'), #Capa conectada
    tf.keras.layers.Dense(units=64, activation='relu'), #Capa conectada
    tf.keras.layers.Dense(units=3) #Capa conectada
])

# Compilación
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1), #Minimización del error
    loss='mean_squared_error'  #predicción a valor real
)


# Entrenamiento
print("Comenzando entrenamiento...") #impreción del mensaje
historial = modelo.fit(anios, np.column_stack((calidad_buena, calidad_regular, calidad_mala)), epochs=300, verbose=False) #Entrenamiento de modelos
print("Modelo entrenado!") #impreción del mensaje

# Guardar el modelo en formato HDF5
modelo.save('modelo_calidad_aire_5.h5') #guardar el modelo

# Convertir a TensorFlow.js (opcional)
!tensorflowjs_converter --input_format keras modelo_calidad_aire_5.h5 calidad_aire_5 #Conversión del modelo
```
